```python
from sklearn.preprocessing import OneHotEncoder
import numpy as np

# Columns that are already numeric (int64) — no conversion needed
# Year, Vehicle_Age, Is_AMG, Is_4MATIC, Mileage_Miles

# Columns that are object (string) — need OneHotEncoding
cat_cols = ['Model_Series', 'Trim_Level', 'Body_Type', 'Mileage_Category']

encoder = OneHotEncoder(sparse_output=False, drop='first')
encoded_array = encoder.fit_transform(df[cat_cols])
encoded_df = pd.DataFrame(encoded_array, columns=encoder.get_feature_names_out(cat_cols))
encoded_df = encoded_df.astype(int)

# Numeric columns stay as-is
num_cols = ['Year', 'Vehicle_Age', 'Is_AMG', 'Is_4MATIC', 'Mileage_Miles']
num_df = df[num_cols].reset_index(drop=True)

# Combine numeric + encoded into final feature set (all int)
x = pd.concat([num_df, encoded_df], axis=1)
y = df['Price_USD']

print("All X dtypes:\n", x.dtypes)
print(f"\nX shape: {x.shape}")
print(f"y shape: {y.shape}")
print(f"\nX columns: {x.columns.tolist()}")
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=42,test_size=0.8)
```
